```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms
import os
```

```python
train_path = "../data/train/original/"
mask_path = "../data/test/"
```

```python
def image_show(path, transform=None):
    img = Image.open(path)
    if transform:
        img = transform(img)
    plt.imshow(img)
    plt.show()
```

```python
image_path = os.path.join(train_path, "leaf01.jpg")
image_show(image_path)
```

```python
transform = transforms.RandomHorizontalFlip()
image_show(image_path, transform)
```

```python
transform = transforms.RandomVerticalFlip()
image_show(image_path, transform)
```

```python
transform = transforms.RandomRotation(45)
image_show(image_path, transform)
```
